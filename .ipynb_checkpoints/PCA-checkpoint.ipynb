```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import seaborn as sns
#Suppress scientific notation
np.set_printoptions(suppress=True)
```

```python
#Load the dataset and seperate the target colum
df = pd.read_csv('Absenteeism/Absenteeism_at_work.csv', delimiter=';')
target = df['Absenteeism time in hours']
df = df.drop('Absenteeism time in hours',axis=1)
df
```

<hr />
\begin{align}
x^\prime = \frac{x-mean(x)}{max(x)-min(x)}
\end{align}

```python
def mean_norm(X):
    return ((X - np.mean(X)) / (np.max(X) - np.min(X)))
```

```python
#Apply Mean Normalisation to the dataset
df_n = mean_norm(df.copy())
df_n
```

\begin{align}
Cov(x)=\frac{1}{m}XX^T
\end{align}

```python
#Calcuate the covariance matrix
def my_cov(X):
    return (np.dot(X,X.T)/df.shape[1])
Xt = df_n.T
covariance_matrix=my_cov(Xt)
```

```python
#Compute the Eigen values and vectors
U,S,V = np.linalg.svd(covariance_matrix)
```

```python
#Select the eigen values which give at least 90% of the variances
threshold = 0.9
K=1
while True:
    preserve = np.sum(S[0:K])/np.sum(S)
    if (preserve >= threshold):
        break
    K+=1

print(str(df.shape[1]) + ' features to be reduced to ' + str(K) + ' principle components while keeping ' \
      + '{0:.0%}'.format(preserve) + ' of the information.')
```

```python
#Compute the Principle Components
pca_x = np.dot(df,V)
#Drop the unwanted components
pca_x = pca_x[:,0:K]
pca_x.shape
```

```python
#The model can now be applied to the Principle Components
```

```python
#Tried to visualize the data for Principle Components 1 & 2 but I am not sure this is the right way
```

```python
#Visualize PC1 and PC2 against the target
PCplot = pd.DataFrame(pca_x[:,0:2],columns=['PC1','PC2'])
PCplot['target'] = target
sns.lmplot('PC1','PC2',data=PCplot,hue='target',scatter_kws={'s':50},fit_reg=False)
```
