```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import pearsonr
```

<h4><center>Pearson Correlation</center></h4>
\begin{align}
r = \frac{\sum{(x-\bar{x})(y-\bar{y})}}{\sqrt{\sum{(x-\bar{x})^2\sum{(y-\bar{y})^2}}}}
\end{align}

```python
#Define a function for Pearson Correlation from first principles
def my_pearsonr(x,y):
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    numerator=np.sum((x-mean_x)*(y-mean_y))
    denominator=(np.sum((x-mean_x)**2) * np.sum((y-mean_y)**2))**0.5
    return numerator / denominator, None
```

```python
#Load the dataset and seperate the target colum
df = pd.read_csv('Absenteeism/Absenteeism_at_work.csv', delimiter=';')
target = df['Absenteeism time in hours']
df = df.drop('Absenteeism time in hours',axis=1)
df
```

```python
#Calculate the feature correlation
cols = df.columns.size
relations = np.zeros((cols,cols))

for i in range(cols):
    for j in range(i,cols):
        a,_ = my_pearsonr(df.iloc[:,i],df.iloc[:,j])
        b,_ = my_pearsonr(df.iloc[:,j],df.iloc[:,i])
        relations[i,j]=a
        relations[j,i]=b
corr = pd.DataFrame(data=relations,index=df.columns,columns=df.columns)
```

```python
#Check if the correlations tally with pandas correlation function
pd_relations = df.corr(method='pearson').to_numpy()
np.allclose(relations,pd_relations)
```

```python
#Show the correlation heatmap
plt.figure(figsize=(15,10))
sns.heatmap(corr,annot=True,cmap='bwr')
```

```python
#Function to identify which features to drop
def get_correlated_features(corr,treshold):
    feature_list=[]
    cols = corr.columns.size
    print('Correlated Features:')
    for i in range(cols):
        for j in range(i+1,cols):
            if(corr.iloc[i,j]>treshold):
                feature_list.append((i,j))
                print(corr.columns[i] + ' <--> ' + corr.columns[j])
    return np.array(feature_list)

#Select features to drop
treshold=0.6
to_drop = get_correlated_features(corr,treshold)
print()
for x in to_drop[:,0]:
    print('Dropping feature: ' + corr.columns[x])
    df = df.drop(corr.columns[x],axis=1)
df        
```
